# Scrape 1

This block of contains a scrape code to parse information in the filter page of OER. It covers the basic details of courses - and I've also added support for
scraping lesson information one page deeper into the cycle. {The is an updated version of this. So, no need to use it.}

```python
import requests
from bs4 import BeautifulSoup

# Base URL for lessons on OER Commons
base_url = 'https://oercommons.org'

# URL to fetch elementary-level English resources
url = 'https://oercommons.org/search?batch_size=20&sort_by=visits&view_mode=summary&f.general_subject=english-language-arts&f.sublevel=lower-primary'

# Request to fetch the main page with lesson summaries
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# Find all articles containing course information
courses = soup.find_all('article', class_='js-index-item')

# List to store lesson metadata
lesson_data = []

# Loop through each course and extract details
for course in courses:
    try:
        title = course.find('div', class_='item-title').text.strip()
        link = course.find('a', class_='item-link')['href']
        author = course.find('dl').find_all('dd')[0].text.strip()
        subject = course.find('dl').find_all('dd')[1].text.strip()
        material_type = course.find('dl').find_all('dd')[2].text.strip()
        description = course.find('div', class_='abstract').text.strip()
        license_info = course.find('div', class_='cou-bucket').text.strip()

        lesson_url = base_url + link

        # Make a GET request to fetch the individual lesson page
        lesson_response = requests.get(lesson_url)
        lesson_soup = BeautifulSoup(lesson_response.content, 'html.parser')

        # Extracting additional metadata
        keywords = lesson_soup.find('meta', {'itemprop': 'keywords'})['content'] if lesson_soup.find('meta', {'itemprop': 'keywords'}) else 'N/A'
        date_created = lesson_soup.find('meta', {'itemprop': 'dateCreated'})['content'] if lesson_soup.find('meta', {'itemprop': 'dateCreated'}) else 'N/A'
        lesson_title = lesson_soup.find('h1', class_='material-title').a.text if lesson_soup.find('h1', class_='material-title') else title
        resource_url = lesson_soup.find('a', {'id': 'goto'})['href'] if lesson_soup.find('a', {'id': 'goto'}) else 'N/A'

        # Add lesson data to the list
        lesson_data.append({
            'title': lesson_title,
            'author': author,
            'subject': subject,
            'material_type': material_type,
            'description': description,
            'license_info': license_info,
            'keywords': keywords,
            'date_created': date_created,
            'resource_url': resource_url,
        })

    except Exception as e:
        print(f"Error occurred while processing {title}: {e}")

# Print the extracted lesson metadata
for lesson in lesson_data:
    print(f"Title: {lesson['title']}")
    print(f"Author: {lesson['author']}")
    print(f"Subject: {lesson['subject']}")
    print(f"Material Type: {lesson['material_type']}")
    print(f"Description: {lesson['description']}")
    # print(f"License: {lesson['license_info']}")
    print(f"Keywords: {lesson['keywords']}")
    print(f"Date Created: {lesson['date_created']}")
    print(f"Resource URL: {lesson['resource_url']}")
    print("________________________________\n\n")
```

```python
import requests
from bs4 import BeautifulSoup

# Function to scrape detailed lesson information for /courseware/lesson/ resources
def scrape_lesson_info(resource_url):
    response = requests.get(resource_url)
    soup = BeautifulSoup(response.content, 'html.parser')

    lesson_info = {
        "lesson_title": "N/A",
        "overview": "N/A",
        "objectives": [],
        "activity_details": "N/A",
        "author": "N/A",
        "subject": "N/A",
        "material_type": "N/A",
        "level": "N/A",
        "tags": [],
        "license": "N/A",
        "media_formats": "N/A",
        "related_resources": []
    }

    # Scrape lesson title
    title_tag = soup.find('h1', class_='lesson-title')
    if title_tag:
        lesson_info["lesson_title"] = title_tag.text.strip()

    # Scrape overview
    overview_tag = soup.find('h2', class_='lesson-subtitle', id='overview')
    if overview_tag:
        overview_content = overview_tag.find_next('div', class_='formatted-text')
        if overview_content:
            lesson_info["overview"] = overview_content.text.strip()

    # Scrape objectives
    objectives_tag = soup.find('h1', class_='lesson-task-title', id='task1')
    if objectives_tag:
        objectives_list = objectives_tag.find_next('ul')
        if objectives_list:
            objectives = objectives_list.find_all('li')
            lesson_info["objectives"] = [obj.text.strip() for obj in objectives]

    # Scrape activity details
    activity_tag = soup.find('h1', class_='lesson-task-title', id='task2')
    if activity_tag:
        activity_content = activity_tag.find_next('div', class_='formatted-text')
        if activity_content:
            lesson_info["activity_details"] = activity_content.text.strip()

    # Scrape author information
    author_tag = soup.find('dt', string='Author:')
    if author_tag:
        authors = author_tag.find_next('dd').find_all('a')
        lesson_info['author'] = ', '.join([author.text.strip() for author in authors])

    # Scrape subject information
    subject_tag = soup.find('dt', string='Subject:')
    if subject_tag:
        lesson_info['subject'] = subject_tag.find_next('dd').text.strip()

    # Scrape material type
    material_type_tag = soup.find('dt', string='Material Type:')
    if material_type_tag:
        lesson_info['material_type'] = material_type_tag.find_next('dd').text.strip()

    # Scrape level
    level_tag = soup.find('dt', string='Level:')
    if level_tag:
        lesson_info['level'] = level_tag.find_next('dd').text.strip()

    # Scrape tags
    tags_tag = soup.find('dd', {'id': 'js-tags-list'})
    if tags_tag:
        tags = tags_tag.find_all('a')
        lesson_info['tags'] = [tag.text.strip() for tag in tags]

    # Scrape license
    license_tag = soup.find('dt', string='License:')
    if license_tag:
        lesson_info['license'] = license_tag.find_next('dd').find('a').text.strip()

    # Scrape media formats
    media_formats_tag = soup.find('dt', string='Media Formats:')
    if media_formats_tag:
        lesson_info['media_formats'] = media_formats_tag.find_next('dd').text.strip()

    # Scrape related resources
    related_resources = soup.find_all('article', class_='related-resource')
    for resource in related_resources:
        resource_title = resource.find('h1', class_='related-resource-title').text.strip()
        view_link_tag = resource.find('a', {'target': '_blank'})
        download_link_tag = resource.find('a', class_='mod-download')

        resource_info = {
            "title": resource_title,
            "view_link": view_link_tag.get('href') if view_link_tag else 'N/A',
            "download_link": download_link_tag.get('href') if download_link_tag else 'N/A'
        }
        lesson_info['related_resources'].append(resource_info)

    return lesson_info


# Function to scrape detailed content for /authoring/ resources
def extract_oer_content(url):
    res = requests.get(url)
    soup = BeautifulSoup(res.text, 'html.parser')

    summary_sec = soup.find('div', {'id': 'summary-tab'})
    summary = summary_sec.get_text(strip=True) if summary_sec else None

    learning_goals_sec = summary_sec.find('ul') if summary_sec else None
    learning_goals = [li.get_text(strip=True) for li in learning_goals_sec.find_all('li')] if learning_goals_sec else []

    toc_sec = soup.find('div', {'id': 'toc-tab'})
    toc_headers = [header.get_text(strip=True) for header in toc_sec.find_all('div', {'class': 'header'})] if toc_sec else []

    videos = []
    for video_object in soup.find_all('div', {'itemtype': 'schema.org/VideoObject'}):
        figure_tag = video_object.find('figure', {'class': 'embed video align-left oembed'})
        if figure_tag and figure_tag.has_attr('data-url'):
            video_url = figure_tag['data-url']
            videos.append(video_url)

    article = soup.find('article', {'class': 'authored-content'})
    title = article.find('h1', {'class': 'authoring-title'}).get_text(strip=True) if article else None
    created_by = article.find('span', {'class': 'content-author'}).get_text(strip=True) if article else None

    return {
        "Summary": summary,
        "Learning Goals": learning_goals,
        "TOC Headers": toc_headers,
        "Videos": videos,
        "Title": title,
        "Created by": created_by
    }


# Base URL for lessons on OER Commons
base_url = 'https://oercommons.org'

# URL to fetch elementary-level English resources
url = 'https://oercommons.org/search?batch_size=20&sort_by=visits&view_mode=summary&f.general_subject=english-language-arts&f.sublevel=lower-primary'

# Request to fetch the main page with lesson summaries
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# Find all articles containing course information
courses = soup.find_all('article', class_='js-index-item')

# List to store lesson metadata
lesson_data = []


# Loop through each course and extract details
for course in courses:
    try:
        title = course.find('div', class_='item-title').text.strip()
        link = course.find('a', class_='item-link')['href']
        author = course.find('dl').find_all('dd')[0].text.strip()
        subject = course.find('dl').find_all('dd')[1].text.strip()
        material_type = course.find('dl').find_all('dd')[2].text.strip()
        description = course.find('div', class_='abstract').text.strip()
        license_info = course.find('div', class_='cou-bucket').text.strip()

        lesson_url = base_url + link

        # Make a GET request to fetch the individual lesson page
        lesson_response = requests.get(lesson_url)
        lesson_soup = BeautifulSoup(lesson_response.content, 'html.parser')

        # Extract resource URL
        resource_url = lesson_soup.find('a', {'id': 'goto'})['href'] if lesson_soup.find('a', {'id': 'goto'}) else 'N/A'

        # Check if the resource_url contains '/courseware/lesson/' or '/authoring/'
        if '/courseware/lesson/' in resource_url:
            detailed_info = scrape_lesson_info(resource_url)
        elif '/authoring/' in resource_url:
            detailed_info = extract_oer_content(resource_url)
        else:
            detailed_info = {}

        # Combine both basic and detailed lesson data
        combined_data = {
            'title': title,
            'author': author,
            'subject': subject,
            'material_type': material_type,
            'description': description,
            'license_info': license_info,
            'resource_url': resource_url,
        }

        # Add detailed info to the combined_data, removing duplicates
        combined_data.update({k: v for k, v in detailed_info.items() if v and k not in combined_data})

        # Add to final data list
        lesson_data.append(combined_data)

    except Exception as e:
        print(f"Error occurred while processing {title}: {e}")

# Print the extracted lesson metadata
for lesson in lesson_data:
    print("Lesson Details:")
    for key, value in lesson.items():
        print(f"{key}: {value}")

    print("________________________________\n\n")
    print("________________________________\n\n")
```

```python
import requests
from bs4 import BeautifulSoup
from pydantic import BaseModel, HttpUrl
from typing import List, Optional
import json

# Pydantic models for lesson information
class RelatedResource(BaseModel):
    title: str
    view_link: Optional[HttpUrl] = "N/A"
    download_link: Optional[HttpUrl] = "N/A"


class LessonInfo(BaseModel):
    lesson_title: str = "N/A"
    overview: str = "N/A"
    objectives: List[str] = []
    activity_details: str = "N/A"
    author: str = "N/A"
    subject: str = "N/A"
    material_type: str = "N/A"
    level: str = "N/A"
    tags: List[str] = []
    license: str = "N/A"
    media_formats: str = "N/A"
    related_resources: List[RelatedResource] = []


class OERContent(BaseModel):
    Summary: Optional[str]
    Learning_Goals: List[str]
    TOC_Headers: List[str]
    Videos: List[str]
    Title: Optional[str]
    Created_by: Optional[str]


class CombinedData(BaseModel):
    title: str
    author: str
    subject: str
    material_type: str
    description: str
    license_info: str
    resource_url: HttpUrl
    lesson_info: Optional[LessonInfo] = None
    oer_content: Optional[OERContent] = None


# Function to scrape detailed lesson information for /courseware/lesson/ resources
def scrape_lesson_info(resource_url) -> LessonInfo:
    response = requests.get(resource_url)
    soup = BeautifulSoup(response.content, 'html.parser')

    lesson_info = LessonInfo()

    # Scrape lesson title
    title_tag = soup.find('h1', class_='lesson-title')
    if title_tag:
        lesson_info.lesson_title = title_tag.text.strip()

    # Scrape overview
    overview_tag = soup.find('h2', class_='lesson-subtitle', id='overview')
    if overview_tag:
        overview_content = overview_tag.find_next('div', class_='formatted-text')
        if overview_content:
            lesson_info.overview = overview_content.text.strip()

    # Scrape objectives
    objectives_tag = soup.find('h1', class_='lesson-task-title', id='task1')
    if objectives_tag:
        objectives_list = objectives_tag.find_next('ul')
        if objectives_list:
            objectives = objectives_list.find_all('li')
            lesson_info.objectives = [obj.text.strip() for obj in objectives]

    # Scrape activity details
    activity_tag = soup.find('h1', class_='lesson-task-title', id='task2')
    if activity_tag:
        activity_content = activity_tag.find_next('div', class_='formatted-text')
        if activity_content:
            lesson_info.activity_details = activity_content.text.strip()

    # Scrape author information
    author_tag = soup.find('dt', string='Author:')
    if author_tag:
        authors = author_tag.find_next('dd').find_all('a')
        lesson_info.author = ', '.join([author.text.strip() for author in authors])

    # Scrape subject information
    subject_tag = soup.find('dt', string='Subject:')
    if subject_tag:
        lesson_info.subject = subject_tag.find_next('dd').text.strip()

    # Scrape material type
    material_type_tag = soup.find('dt', string='Material Type:')
    if material_type_tag:
        lesson_info.material_type = material_type_tag.find_next('dd').text.strip()

    # Scrape level
    level_tag = soup.find('dt', string='Level:')
    if level_tag:
        lesson_info.level = level_tag.find_next('dd').text.strip()

    # Scrape tags
    tags_tag = soup.find('dd', {'id': 'js-tags-list'})
    if tags_tag:
        tags = tags_tag.find_all('a')
        lesson_info.tags = [tag.text.strip() for tag in tags]

    # Scrape license
    license_tag = soup.find('dt', string='License:')
    if license_tag:
        lesson_info.license = license_tag.find_next('dd').find('a').text.strip()

    # Scrape media formats
    media_formats_tag = soup.find('dt', string='Media Formats:')
    if media_formats_tag:
        lesson_info.media_formats = media_formats_tag.find_next('dd').text.strip()

    # Scrape related resources
    related_resources = soup.find_all('article', class_='related-resource')
    for resource in related_resources:
        resource_title = resource.find('h1', class_='related-resource-title').text.strip()
        view_link_tag = resource.find('a', {'target': '_blank'})
        download_link_tag = resource.find('a', class_='mod-download')

        resource_info = RelatedResource(
            title=resource_title,
            view_link=view_link_tag.get('href') if view_link_tag else 'N/A',
            download_link=download_link_tag.get('href') if download_link_tag else 'N/A'
        )
        lesson_info.related_resources.append(resource_info)

    return lesson_info


# Function to scrape detailed content for /authoring/ resources
def extract_oer_content(url) -> OERContent:
    res = requests.get(url)
    soup = BeautifulSoup(res.text, 'html.parser')

    summary_sec = soup.find('div', {'id': 'summary-tab'})
    summary = summary_sec.get_text(strip=True) if summary_sec else None

    learning_goals_sec = summary_sec.find('ul') if summary_sec else None
    learning_goals = [li.get_text(strip=True) for li in learning_goals_sec.find_all('li')] if learning_goals_sec else []

    toc_sec = soup.find('div', {'id': 'toc-tab'})
    toc_headers = [header.get_text(strip=True) for header in toc_sec.find_all('div', {'class': 'header'})] if toc_sec else []

    videos = []
    for video_object in soup.find_all('div', {'itemtype': 'schema.org/VideoObject'}):
        figure_tag = video_object.find('figure', {'class': 'embed video align-left oembed'})
        if figure_tag and figure_tag.has_attr('data-url'):
            video_url = figure_tag['data-url']
            videos.append(video_url)

    article = soup.find('article', {'class': 'authored-content'})
    title = article.find('h1', {'class': 'authoring-title'}).get_text(strip=True) if article else None
    created_by = article.find('span', {'class': 'content-author'}).get_text(strip=True) if article else None

    return OERContent(
        Summary=summary,
        Learning_Goals=learning_goals,
        TOC_Headers=toc_headers,
        Videos=videos,
        Title=title,
        Created_by=created_by
    )


# Base URL for lessons on OER Commons
base_url = 'https://oercommons.org'

# URL to fetch elementary-level English resources
url = 'https://oercommons.org/search?batch_size=20&sort_by=visits&view_mode=summary&f.general_subject=english-language-arts&f.sublevel=lower-primary'

# Request to fetch the main page with lesson summaries
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# Find all articles containing course information
courses = soup.find_all('article', class_='js-index-item')

# List to store lesson metadata
lesson_data = []

# Loop through each course and extract details
for course in courses:
    try:
        title = course.find('div', class_='item-title').text.strip()
        link = course.find('a', class_='item-link')['href']
        author = course.find('dl').find_all('dd')[0].text.strip()
        subject = course.find('dl').find_all('dd')[1].text.strip()
        material_type = course.find('dl').find_all('dd')[2].text.strip()
        description = course.find('div', class_='abstract').text.strip()
        license_info = course.find('div', class_='cou-bucket').text.strip()

        lesson_url = base_url + link

        # GET request to fetch the individual lesson page
        lesson_response = requests.get(lesson_url)
        lesson_soup = BeautifulSoup(lesson_response.content, 'html.parser')

        # Resource URL
        resource_url = lesson_soup.find('a', {'id': 'goto'})['href'] if lesson_soup.find('a', {'id': 'goto'}) else 'N/A'

        # To store combined information
        combined_data = CombinedData(
            title=title,
            author=author,
            subject=subject,
            material_type=material_type,
            description=description,
            license_info=license_info,
            resource_url=resource_url
        )

        # Filtering
        if '/courseware/lesson/' in resource_url:
            combined_data.lesson_info = scrape_lesson_info(resource_url)
        elif '/authoring/' in resource_url:
            combined_data.oer_content = extract_oer_content(resource_url)

        # Add to final data list
        lesson_data.append(combined_data)

    except Exception as e:
        print(f"Error occurred while processing {title}: {e}")

#JSON file
with open('lesson_data.json', 'w') as json_file:
    json_file.write(
        json.dumps([json.loads(lesson.json()) for lesson in lesson_data], indent=4)
    )

print("Data saved to lesson_data.json")
```
